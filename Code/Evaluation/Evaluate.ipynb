```python
import pandas as pd
from sklearn.metrics import precision_score, recall_score, f1_score, accuracy_score
```

```python
def calculate_metrics(true_labels, predicted_labels):
    accuracy = accuracy_score(true_labels, predicted_labels)
    precision = precision_score(true_labels, predicted_labels, average='macro')
    recall = recall_score(true_labels, predicted_labels, average='macro')
    f1 = f1_score(true_labels, predicted_labels, average='macro')
    return accuracy, precision, recall, f1
```

```python
df = pd.read_csv('prediction.csv')
```

```python
prediction_list = df['model_name'].to_list()
```

Post-processing on prediction

```python
def clean(text):
  k = text[:50].lower()
  #k = text
  if '0' in k or 'rejected' in k:
    return 0
  elif '1' in k or 'accepted' in k:
    return 1
  else:
    return 2
```

```python
actual = [int(i) for i in df['Label'].tolist()]
# pred = [clean(i) for i in df['llama_van_pe'].tolist()]
pred = [clean(i) for i in prediction_list]
```

```python
a1 = []
p1 = []
```

```python
for i,e in enumerate(pred):
  if e == 1 or e==0:
    a1.append(actual[i])
    p1.append(e)
```

```python
#len(a1)
```

```python
accuracy, precision, recall, f1 = calculate_metrics(a1, p1)
print("Accuracy:", accuracy)
print("Macro Precision:", precision)
print("Macro Recall:", recall)
print("Macro F1-score:", f1)
```
